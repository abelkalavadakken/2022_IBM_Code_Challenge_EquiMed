```python
import pandas as pd
import numpy as np
import os
import shutil
```

```python
dir = 'preprocessed_images'
files = os.listdir(dir)
```

```python
df = pd.read_csv('full_df.csv')
df.head(3)
```

```python
df.C
```

```python
def if_cataract(text):
    if "cataract" in text:
        return 1
    else:
        return 0
        
```

```python
df['lt_cataract'] = df['Left-Diagnostic Keywords'].apply(lambda x:if_cataract(x))

df['rt_cataract'] = df['Right-Diagnostic Keywords'].apply(lambda x:if_cataract(x))
```

```python
lt_cataract = df.loc[(df.C == 1) & (df.lt_cataract == 1)]['Left-Fundus'].values
print(lt_cataract[:5])
```

```python
rt_cataract = df.loc[(df.C == 1) & (df.rt_cataract == 1)]['Right-Fundus'].values
print(rt_cataract[:5])
```

```python
print('No. of Left Cataract images: ', len(lt_cataract))
print('No. of Right Cataract images: ', len(rt_cataract))
```

```python
left_normal = df.loc[(df.C == 0) & (df.N == 1)]['Left-Fundus'].sample(300).values

left_normal[:10]
```

```python
right_normal = df.loc[(df.C == 0) & (df.N == 1)]['Right-Fundus'].sample(300).values
right_normal[:5]
```

```python
print('No. of Left normal images: ', len(left_normal))
print('No. of Right normal images: ', len(right_normal))
```

```python
cataract = np.concatenate((lt_cataract, rt_cataract), axis=0)
cataract[:10]
```

```python
dest = 'cataract_extracted'
os.mkdir('cataract_extracted')
```

```python
for filename in cataract:
    src_path = os.path.join(dir,filename)
    try:
        shutil.copy(src_path,dest)
        print(src_path)
    except OSError as e:
        print(e.errno)
        continue

    
```

```python
normal = np.concatenate((right_normal, left_normal), axis=0)
normal[:10]
```

```python
dest = 'normal_extracted'
os.mkdir('normal_extracted')
```

```python
for filename in normal:
    src_path = os.path.join(dir,filename)
    try:
        shutil.copy(src_path,dest)
        print(src_path)
    except OSError as e:
        print(e.errno)
        continue
```
