```python
import cv2
import os
import glob
import PIL
import matplotlib.pyplot as  plt
```

```python
inputFolder = 'cataract'
os.mkdir('Resized Catract')
```

```python
i = 0
for img in glob.glob(inputFolder+'/*.png'):
    image = cv2.imread(img)
    crop = image[0:1800, 500:2070]
    scale = 0.3
    width = int(crop.shape[1] * scale)
    height = int(crop.shape[0] * scale)
    dim = (width,height)
    imgResized = cv2.resize(crop, dim,)
    # crop = imgResized[10:200, 100:400]
    cv2.imwrite('Resized Catract/image%i.jpg' %i, imgResized)

    i += 1
```
